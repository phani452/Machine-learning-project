# coreelation matrix

```python
corr_matrix = data.corr()
corr_matrix
```

```python
import seaborn as sns #graphic tool
```

# heatmap

```python
mask = np.zeros_like(corr_matrix)                                              #heatmap with correlation values

f, ax = plt.subplots(figsize=(16,16))
ax = sns.heatmap(corr_matrix, mask=mask, vmax=.3, square=True,annot=True)
```

###### from above heatmap we can infer that 
###### the highly correlated columns are
###### 'bedrooms','bathrooms','sqft_living','view','grade','sqft_above','sqft_basement','sqft_living15

```python
p=pd.DataFrame({'maximum of price':data.groupby(data['grade'])['price'].max()}).reset_index() #grouping grad and mean of prices
```
